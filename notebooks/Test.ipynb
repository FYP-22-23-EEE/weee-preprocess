```python
import pandas as pd
import numpy as np
```

```python
NUMBER_OF_ROWS = 10
TIME_GAP = '125ms'
df = pd.DataFrame({
    'timestamp': pd.date_range('2019-01-01', periods=NUMBER_OF_ROWS, freq=TIME_GAP),
    'A': np.random.randn(NUMBER_OF_ROWS),
    'B': np.random.randn(NUMBER_OF_ROWS),
})
```

```python
df1 = df.copy()
df2 = df.copy()
# assign nan to random rows
df2.loc[df2.sample(frac=0.5).index, 'A'] = np.nan
df2.loc[df2.sample(frac=0.5).index, 'B'] = np.nan
```

```python
df2.groupby(pd.Grouper(key='timestamp', freq='500ms')).agg({
    'timestamp': ['first', 'last'],
    'A': ['count', 'sum', 'mean'],
    'B': ['count', 'sum', 'mean'],
})
```
