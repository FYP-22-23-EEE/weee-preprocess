```python
import pandas as pd
import numpy as np
```

```python
NUMBER_OF_ROWS = 1000
TIME_GAP = '125ms'
df = pd.DataFrame({
    'timestamp': pd.date_range('2019-01-01', periods=NUMBER_OF_ROWS, freq=TIME_GAP),
    'A': np.random.randn(NUMBER_OF_ROWS),
    'B': np.random.randn(NUMBER_OF_ROWS),
})
```

```python
df.head()
```

```python
grouped = df.groupby(pd.Grouper(key='timestamp', freq='2s'))

# create windowed dataframe, with A_mean, A_std, B_mean, B_std columns
windowed = grouped.agg({
    'A': ['mean', 'std'],
    'B': ['mean', 'std'],
    'timestamp': ['first', 'last']
})
```
